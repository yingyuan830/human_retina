```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Volcano plot for eye ( retina vs macula)

# Cones

```python
type = 'Cones'
df = pd.read_csv(f"/data/projects/Human-retina/scripts/dif_edgeR/dif_hg_tissue/{type}.csv")
df=df.dropna()
df['minusLog10FDR'] = -np.log10(df['FDR'])

plt.figure(figsize=(10, 6))
plt.scatter(df['logFC'], df['minusLog10FDR'],
            c=np.where((df['logFC'] > -1) & (df['logFC'] < 1) | (df['minusLog10FDR'] < -np.log10(0.01)), "gray", "red"),  
            alpha=0.5,s=20)

plt.title(f'{type}: retina vs macula (Up {up_count}, Down {down_count})',)
plt.xlabel('Log2 Fold Change')
plt.ylabel('-Log10FDR')


plt.axhline(y=-np.log10(0.01), color='blue', linestyle='--')  
plt.axvline(x=-1, color='green', linestyle='--')  
plt.axvline(x=1, color='green', linestyle='--')  

plt.legend()
plt.show()
```
